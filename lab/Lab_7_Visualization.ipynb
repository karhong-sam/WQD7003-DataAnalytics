## Line Plot / Scatter Plot

**matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)**


The coordinates of the points or line nodes are given by x, y.

The optional parameter fmt is a convenient way for defining basic formatting like color, marker and linestyle. It's a shortcut string notation described in the Notes section below.


```
plot(x, y)        # plot x and y using default line style and color
plot(x, y, 'bo')  # plot x and y using blue circle markers
plot(y)           # plot y using x as index array 0..N-1
plot(y, 'r+')     # ditto, but with red plusses
```

You can use Line2D properties as keyword arguments for more control on the appearance. Line properties and fmt can be mixed. The following two calls yield identical results:



```
plot(x, y, 'go--', linewidth=2, markersize=12)
plot(x, y, color='green', marker='o', linestyle='dashed',
     linewidth=2, markersize=12)
```


When conflicting with fmt, keyword arguments take precedence.

Source - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

````python
## Plot scatter plot with marker size = 20, a list of color

x = np.random.rand(20)
y = np.random.rand(20)

plt.scatter(x, y, s=40, c=np.arange(1,21))
plt.colorbar()
````

````python
## Plot scatter plot with marker size = 20, a list of color 
## using colormap of hsv 
## https://matplotlib.org/stable/tutorials/colors/colormaps.html 

x = np.random.rand(20)
y = np.random.rand(20)

plt.scatter(x, y, s=20, c=np.arange(1,21), cmap='Greys')
plt.colorbar()

## try with other colormap, refers to https://matplotlib.org/stable/tutorials/colors/colormaps.html 
````

````python
## Plot scatter plot with marker size = 20, a list of color 
## using colormap of hsv, set the min and max of the color mapping

x = np.random.rand(20)
y = np.random.rand(20)

plt.scatter(x, y, s=20, c=np.arange(1,21), cmap='hsv', vmin=5, vmax=10)
plt.colorbar()
````

````python
# plot multiple bar plot - two groups

import matplotlib.pyplot as plt

courses = ['C', 'C++', 'Java', 'Python']
class_A = [20, 15, 30, 35]
class_B = [51, 21, 22, 11]

X_axis = np.arange(len(courses))

plt.bar(X_axis - 0.2, class_A, width=0.4, label = 'class_A')
plt.bar(X_axis + 0.2, class_B, width=0.4, label = 'class_B')

plt.xticks(X_axis, courses)  #rename the xtick 
plt.legend()
````

````python
# plot multiple bar plot - three groups

import matplotlib.pyplot as plt

courses = ['C', 'C++', 'Java', 'Python']
class_A = [20, 15, 30, 35]
class_B = [51, 21, 22, 11]
class_C = [11, 22, 52, 31]

X_axis = np.arange(len(courses))

plt.bar(X_axis - 0.2, class_A, width=0.2, label = 'class_A')
plt.bar(X_axis      , class_B, width=0.2, label = 'class_B')
plt.bar(X_axis + 0.2, class_C, width=0.2, label = 'class_C')

plt.xticks(X_axis, courses)  #rename the xtick 
plt.legend()
````

````python
# plot multiple bar plot - four groups

import matplotlib.pyplot as plt

courses = ['C', 'C++', 'Java', 'Python']
class_A = [20, 15, 30, 35]
class_B = [51, 21, 22, 11]
class_C = [11, 22, 52, 31]
class_D = [31, 27, 12, 55]

X_axis = np.arange(len(courses))

plt.bar(X_axis - 0.3, class_A, width=0.2, label = 'class_A')
plt.bar(X_axis - 0.1, class_B, width=0.2, label = 'class_B')
plt.bar(X_axis + 0.1, class_C, width=0.2, label = 'class_C')
plt.bar(X_axis + 0.3, class_D, width=0.2, label = 'class_D')

plt.xticks(X_axis, courses)  #rename the xtick 
plt.legend()
````

````python
# plot multiple horizontal bar plot - two groups with error bars

import matplotlib.pyplot as plt

courses = ['C', 'C++', 'Java', 'Python']
class_A = [20, 15, 30, 35]
class_B = [51, 21, 22, 11]
AStd = [2, 3, 4, 1]
BStd = [3, 5, 2, 3]

X_axis = np.arange(len(courses))

plt.bar(X_axis - 0.2, class_A, width=0.4, label = 'class_A', yerr = AStd)
plt.bar(X_axis + 0.2, class_B, width=0.4, label = 'class_B', yerr = BStd)

plt.xticks(X_axis, courses)  #rename the xtick 
plt.legend()
````

## Pie Chart

Source - https://matplotlib.org/devdocs/gallery/pie_and_polar_charts/pie_features.html?highlight=pie%20chart

````python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [10, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes,  explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
````

````python
# plot the frequency / count of female and male in the datasets. 
# Your solution here
gender = pd.unique(data.gender)
count_gender = data['gender'].value_counts()
men_count = count_gender['M']
women_count = count_gender['W']

plt.bar(gender, count_gender.values)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()
````

````python
#@title Solution1
X = data.gender.unique()

gender_count = data['gender'].value_counts()
men_count = gender_count['M']
women_count = gender_count['W']

gender_count.values
X_loc = np.arange(2)

plt.bar(X_loc, gender_count.values, width=0.4)

plt.xticks(X_loc, X)  #rename the xtick 
plt.legend()
plt.xlabel('Gender')
plt.ylabel('Frequency')
````

````python
# plot the average age for female and male in the datasets with an error bar showing the standard deviation
# Your solution here
gender = pd.unique(data.gender)
avg_age = data.groupby('gender').mean()
std_age= data.groupby('gender').std()

men_mean, men_std = avg_age['age']['M'], std_age['age']['M']
women_mean, women_std = avg_age['age']['W'], std_age['age']['W']

mean_age = [men_mean, women_mean]
sd_age = [men_std, women_std]

plt.bar(gender, mean_age, yerr=sd_age)
plt.xlabel('Gender')
plt.ylabel('Average Age')
````
